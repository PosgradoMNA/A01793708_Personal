*   **Pregunta 1.** 

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario. 

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
for line in range(len(docs)):
    docs[line] = re.sub(r'\n$', '', docs[line])      #Quitamos saltos de línea
    docs[line] = re.sub(r'^"|"$', '', docs[line])    #Quitamos comillas dobles al inicio y fin
    docs[line] = re.sub(r'\.$', '', docs[line])       #Quitamos puntos al final
```

```python
tokens = []

for line in docs:
    for word in re.split(r'\s', line):                     #Separamos palabras por caracteres de espacios
        tokens.append(word)
```

```python
dict = {}
query = []

for word in tokens:
    temp = []
    temp = re.findall(r'\w*(?:!{2,})$', word)
    if len(temp) != 0:
        if temp[0] in dict:
            dict[temp[0]] += 1
        else:
            dict[temp[0]] = 1

print('Hay en total: ', len(dict), ' palabras.')
dict.keys()

    
```

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
for line in range(len(docs)):
    docs[line] = re.sub(r'[^a-zA-Z ]', '', docs[line])  #Quitamos lo que no sea alfabetico, guiones bajos y números
    docs[line] = re.sub(r'\s+', ' ', docs[line])        #Quitamos espacios dobles que pudiesen haber
    
docs = [str.lower(line) for line in docs]               #Convertimos a minúsculas
```
